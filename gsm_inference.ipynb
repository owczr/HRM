```python
from hrm_sdk import HRMInferenceClient

client = HRMInferenceClient("http://localhost:8000")
result = client.infer([0, 1, 2, 3])

print(result.predictions)
```

```python
import numpy as np


data = np.load("data/questbench-gsm/test/all__inputs.npy")
labels = np.load("data/questbench-gsm/test/all__labels.npy")
```

```python
data[0]
```

```python
result = client.infer(data[200])
print(result.predictions)
```

```python
labels[200] == result.predictions
```

```python
import tqdm

total_accuracy = 0
exact_accuracy = 0

for X, y in tqdm.tqdm(zip(data, labels)):
    y_pred = client.infer(X).predictions
    accuracy = np.sum(y == y_pred) / len(y_pred)
    total_accuracy += accuracy
    exact_accuracy += int(all(y == y_pred))
```

```python
total_accuracy / data.shape[0]
```

```python
exact_accuracy / data.shape[0]
```
