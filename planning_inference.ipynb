```python
import numpy as np


data = np.load("data/questbench-planning/test/all__inputs.npy")
labels = np.load("data/questbench-planning/test/all__labels.npy")
```

```python
import numpy as np
import tqdm
import itertools


total_accuracy = 0
exact_accuracy = 0
pad_id = 0
divisor = 0

for X, y in tqdm.tqdm(itertools.islice(zip(data, labels), 100)):
    y = np.array(y)
    mask = y != pad_id

    y_pred = np.array(client.infer(X).predictions)

    valid_y = y[mask]
    valid_pred = y_pred[mask]

    accuracy = np.mean(valid_y == valid_pred) if len(valid_y) > 0 else 0.0

    exact = np.all(valid_y == valid_pred) if len(valid_y) > 0 else False

    total_accuracy += accuracy
    exact_accuracy += int(exact)
    divisor += 1

total_accuracy = total_accuracy / divisor
exact_accuracy = exact_accuracy / divisor


print(f"{total_accuracy = }")
print(f"{exact_accuracy = }")
```
